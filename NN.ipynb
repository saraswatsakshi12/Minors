```python
# we are designing a NN with Two layers (input / output) and one hidden layer
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Read the data from MNIST dataset
(X_train,y_train),(X_test,y_test)  = keras .datasets.mnist.load_data()
```

```python
# we have to convert 28 X 28 image into a single dimesional array having number of 784 elements

X_train.shape
```

```python
X_train = X_train /255
X_test = X_test / 255
```

```python
# Now flattern the image

X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_test_flattened.shape
```

```python
# NOW WE ARE CREATING A SIMPLE NN , INPUT LAYER with 28*28 = 784 elements and output layer with 10 elements (0 to 9)
```

```python
model = keras.Sequential([
                  keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')])

#Sequesntial means we are having a stack of layers in NN, In stack every layer as one element ,
#dense means one layer is connected with all elments of II Layer

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
     )

# read the detail of different optimizer , it alow globle optima in efficent way

# we can use mean squre error or any other in loss , read gradient decent or cost function

model.fit(X_train_flattened, y_train, epochs=5)   # epochs are number of iteration in NN
```

```python
# If you dont find not good accrecy like (40-50%) not like this (85 to 89 %) then you have to 'scaled' , so scaled it
# means scale the whole array (divide it by 255)  means X_train = X_train /255 and X_test=X_test / 255 before flatterned
```

```python
# Now evalute the accurecy with test data set
model.evaluate(X_test_flattened, y_test)
```

```python
# Add one hidden layer (Generaly improve the performance) by just copy paste the previous code and with minor change


model = keras.Sequential([
                  keras.layers.Dense(100, input_shape=(784,),activation='relu') ,
                  keras.layers.Dense(100, input_shape=(784,),activation='relu'),                 # 100 shows the No of neurons there are NOT certain rules for decision
                  keras.layers.Dense(10,activation='sigmoid')                                        # last layer doent need input shape becoz it is alredy decided in I layer
                  ])


model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
     )


model.fit(X_train_flattened, y_train, epochs=5)
```

optimizer, hidden layer, sign language recognition system

```python
# Now again evalute the accurecy with test data set
model.evaluate(X_test_flattened, y_test)
```

```python
## One importent Note:
# Keras provide a fuction to Flattern Not by 'X_train_flattened' array

model = keras.Sequential([
                  keras.layers.Flatten(input_shape=(28,28)),
                  keras.layers.Dense(100, activation='relu') ,
                  keras.layers.Dense(10,activation='sigmoid')
                  ])


model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
     )


model.fit(X_train, y_train, epochs=5)
```
